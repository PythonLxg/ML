```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import export_graphviz
import graphviz
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_moons
import mglearn
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.datasets import make_blobs


cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=42)
tree_clf = DecisionTreeClassifier(random_state=0)
tree_clf.fit(X_train, y_train)
print('train set score:{:.3f}'.format(tree_clf.score(X_train, y_train)))
print('test set score:{:.3f}'.format(tree_clf.score(X_test, y_test)))
```

```python
tree_clf1 = DecisionTreeClassifier(max_depth=4, random_state=0)
tree_clf1.fit(X_train, y_train)
print('train set score:{:.3f}'.format(tree_clf1.score(X_train, y_train)))
print('test set score:{:.3f}'.format(tree_clf1.score(X_test, y_test)))
```

```python
export_graphviz(tree_clf1, out_file='tree_clf1.dot', class_names=['maligant', 'benign'],
                feature_names=cancer.feature_names, impurity=False, filled=True)

with open('tree_clf1.dot') as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

```python
print(cancer.feature_names)
print('\nFeature importances:\n{}'.format(tree_clf1.feature_importances_))  # 树的特征重要性
```

```python
def plot_feature_importances_cancer(model):
    n_features = cancer.data.shape[1]
    plt.barh(range(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), cancer.feature_names)
    plt.xlabel('Feature importance')
    plt.ylabel('Feature')
    
plot_feature_importances_cancer(tree_clf1)
```

```python
X, y = make_moons(n_samples=100, noise=.25, random_state=3)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

forest_clf = RandomForestClassifier(n_estimators=5, random_state=2)
forest_clf.fit(X_train, y_train)

fig, axes = plt.subplots(2, 3, figsize=(20, 10))
for i, (ax, tree) in enumerate(zip(axes.ravel(), forest_clf.estimators_)):
    ax.set_title('Tree {}'.format(i))
    mglearn.plots.plot_tree_partition(X_train, y_train, tree, ax=ax)
     
mglearn.plots.plot_2d_separator(forest_clf, X_train, fill=True, ax=axes[-1, -1])

axes[-1, -1].set_title('Random Forest')
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train)
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0)
forest_clf1 = RandomForestClassifier(n_estimators=100, random_state=0)
forest_clf1.fit(X_train, y_train)

print('Training set score:{}'.format(forest_clf1.score(X_train, y_train)))
print('Test set score:{}'.format(forest_clf1.score(X_test, y_test)))
```

```python
X, y = make_blobs(centers=4, random_state=8)
y = y % 2

mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel('Feature 0')
plt.ylabel('Feature 1')
```

```python
from mpl_toolkits.mplot3d import Axes3D, axes3d
from sklearn.svm import LinearSVC

X_new = np.hstack([X, X[:, 1:] ** 2])  # 特征多项式

linear_svm_3d = LinearSVC().fit(X_new, y)
coef, intercept = linear_svm_3d.coef_.ravel(), linear_svm_3d.intercept_

# 显示决策边界
figure = plt.figure()
ax = Axes3D(figure, elev=-152, azim=-26)
xx = np.linspace(X_new[:, 0].min() - 2, X_new[:, 0].max() + 2, 50)
yy = np.linspace(X_new[:, 1].min() - 2, X_new[:, 1].max() + 2, 50)

XX, YY = np.meshgrid(xx, yy)
ZZ = (coef[0] * XX + coef[1] * YY + intercept) / -coef[2]
ax.plot_surface(XX, YY, ZZ, rstride=8, cstride=8, alpha=.3)
mask = y == 0  # 首先画出y==0的点，再画出y==1的点
ax.scatter(X_new[mask, 0], X_new[mask,1], X_new[mask, 2], marker='v',
          c='b', cmap=mglearn.cm2, s=60)
ax.scatter(X_new[~mask, 0], X_new[~mask,1], X_new[~mask, 2], marker='^',
          c='r', cmap=mglearn.cm2, s=60)

ax.set_xlabel('feature0')
ax.set_ylabel('feature1')
ax.set_zlabel('feature1 ** 2')
```

```python
ZZ = YY ** 2
dec = linear_svm_3d.decision_function(np.c_[XX.ravel(), YY.ravel(), ZZ.ravel()])
plt.contourf(XX, YY, dec.reshape(XX.shape), levels=[dec.min(), 0, dec.max()], 
             cmap=mglearn.cm2, alpha=.5)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel('Feature 0')
plt.ylabel('Feature 1')
```

```python
from sklearn.svm import SVC

X, y = mglearn.tools.make_handcrafted_dataset()
svm = SVC(kernel='rbf', C=10, gamma=.1).fit(X, y)
mglearn.plots.plot_2d_separator(svm, X, eps=.5)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
sv = svm.support_vectors_
sv_labels = svm.dual_coef_.ravel() > 0
mglearn.discrete_scatter(sv[:, 0], sv[:, 1], sv_labels, s=15, markeredgewidth=3)
plt.xlabel('Feature 0')
plt.ylabel('Feature 1')
```

```python
fig, axes = plt.subplots(3, 3, figsize=(15, 10))

for ax, C in zip(axes, [-1, 0, 3]):
    for a, gamma in zip(ax, range(-1, 2)):
        mglearn.plots.plot_svm(log_C=C, log_gamma=gamma, ax=a)
        
axes[0, 0].legend(['calss 0', 'class 1', 'sv class 0', 'sv class 1'], ncol=4, loc=(.9, 1.2))
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0)
svc = SVC()
svc.fit(X_train, y_train)

print('Training set score:{:.4f}'.format(svc.score(X_train, y_train)))
print('Test set score:{:.4f}'.format(svc.score(X_test, y_test)))
```

```python
# 查看特征的变化范围
plt.plot(X_train.min(axis=0), 'o', label='min')
plt.plot(X_train.max(axis=0), '^', label='max')
plt.legend()
plt.xlabel('Feature index')
plt.ylabel('Feature magnitude')
plt.yscale('log')
```

```python
# 手工进行数据缩放
min_on_training = X_train.min(axis=0)
max_on_training = X_train.max(axis=0)
range_on_training = max_on_training - min_on_training

X_train_scaled = (X_train - min_on_training) / range_on_training
print('Minimum for each feature:\n{}'.format(X_train_scaled.min(axis=0)))
print('Maximum for each feature:\n{}'.format(X_train_scaled.max(axis=0)))
```

```python
X_test_scaled = (X_test - min_on_training) / range_on_training
svc = SVC()
svc.fit(X_train_scaled, y_train)

print('Training set score:{:.4f}'.format(svc.score(X_train_scaled, y_train)))
print('Test set score:{:.4f}'.format(svc.score(X_test_scaled, y_test)))
```

```python
svc = SVC(C=1000)
svc.fit(X_train_scaled, y_train)

print('Training set score:{:.4f}'.format(svc.score(X_train_scaled, y_train)))
print('Test set score:{:.4f}'.format(svc.score(X_test_scaled, y_test)))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.datasets import make_circles

X, y = make_circles(noise=.25, factor=.5, random_state=1)
y_named = np.array(['blue', 'red'])[y]

X_train, X_test, y_train_named, y_test_named, y_train, y_test = train_test_split(X, y_named, y, random_state=0)

gb_clf = GradientBoostingClassifier(random_state=0)
gb_clf.fit(X_train, y_train_named)

print('X_test.shape:{}'.format(X_test.shape))
print('Decision function shape:{}'.format(gb_clf.decision_function(X_test).shape))  # 分类器的不确定度估计 decision_function
print('Decision function:{}'.format(gb_clf.decision_function(X_test)[:6]))  # 前几个元素的置信程度
print('\nPredict_proba shape:{}'.format(gb_clf.predict_proba(X_test).shape))  # 分类器的不确定度估计 predict_proba
print('Predict_proba:\n{}'.format(gb_clf.predict_proba(X_test)[:6]))  # 前几个元素的类别概率
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))
mglearn.tools.plot_2d_separator(gb_clf, X, ax=axes[0], alpha=.6, 
                                fill=True, cm=mglearn.cm2)
scores_image = mglearn.tools.plot_2d_scores(gb_clf, X, ax=axes[1], 
                                            alpha=.6, cm=mglearn.ReBl)

for ax in axes:
    mglearn.discrete_scatter(X_test[:, 0], X_test[:, 1], y_test,
                             markers='^', ax=ax)
    mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train,
                             markers='o', ax=ax)
    ax.set_xlabel('Feaure 0')
    ax.set_ylabel('Feature 1')
cbar = plt.colorbar(scores_image, ax=axes.tolist())
axes[0].legend(['Test class 0', 'Test class 1', 'Train class 0',
                'Train class 1'], ncol=4, loc=(.3, 1.1))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))
mglearn.tools.plot_2d_separator(gb_clf, X, ax=axes[0], alpha=.6, 
                                fill=True, cm=mglearn.cm2)
scores_image = mglearn.tools.plot_2d_scores(gb_clf, X, ax=axes[1], 
                                            alpha=.6, cm=mglearn.ReBl, function='predict_proba')  # 边界更加明确，清晰

for ax in axes:
    mglearn.discrete_scatter(X_test[:, 0], X_test[:, 1], y_test,
                             markers='^', ax=ax)
    mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train,
                             markers='o', ax=ax)
    ax.set_xlabel('Feaure 0')
    ax.set_ylabel('Feature 1')
cbar = plt.colorbar(scores_image, ax=axes.tolist())
axes[0].legend(['Test class 0', 'Test class 1', 'Train class 0',
                'Train class 1'], ncol=4, loc=(.3, 1.1))
```

```python
from sklearn.datasets import load_iris
# 多分类问题的不确定度

iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state=42)

gb_clf = GradientBoostingClassifier(learning_rate=0.01, random_state=0)
gb_clf.fit(X_train, y_train)

print('Decision function shape:{}'.format(gb_clf.decision_function(X_test).shape))
print('Decision function:\n{}'.format(gb_clf.decision_function(X_test)[:6, :]))
print('\nPredict_proba shape:{}'.format(gb_clf.predict_proba(X_test).shape))
print('Predict_proba:\n{}'.format(gb_clf.predict_proba(X_test)[:6, :]))
```

```python
print('Argmax of decision function:\n{}'.format(np.argmax(gb_clf.decision_function(X_test), axis=1)))
print('Argmax of predict_proba:\n{}'.format(np.argmax(gb_clf.predict_proba(X_test), axis=1)))
print('Predictions:\n{}'.format(gb_clf.predict(X_test)))
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()

named_target = iris.target_names[y_train]
logreg.fit(X_train, named_target)
print('unique calss in training data:\n{}'.format(logreg.classes_))
print('predictions:\n{}'.format(logreg.predict(X_test)[:10]))
argmax_dec_func = np.argmax(logreg.decision_function(X_test), axis=1)
print('argmax of decision function:\n{}'.format(argmax_dec_func[:10]))
print('argmax combined with classed_:\n{}'.format(logreg.classes_[argmax_dec_func][:10]))
```
